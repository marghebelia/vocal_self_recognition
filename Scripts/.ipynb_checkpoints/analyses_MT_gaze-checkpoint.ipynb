### SOABB1 - Video analysis of infant looking time
Here we want to know the % of time spent looking in the mirror in phases 1 (before the mark) and 2 (after the mark)

Goals:
1. Find the average looking time (in seconds, + standard deviation) towards the mirror in phase 1 and phase 2.
2. Difference in percentages of time spent looking at the mirror in phase 1 and phase 2

```python
### IMPORT PACKAGES
import matplotlib.pylab as plt
import numpy as np
import os, glob
import pandas
import seaborn as sns
import re
from scipy import stats
import statsmodels.api as sm
from statsmodels.formula.api import ols

#pandas.options.display.max_columns = 1000
#pandas.options.display.max_rows = 2000
```

```python
### OPEN FILE CONTAINING F0 VALUES
resFile = DIR + current_file
df = pandas.read_csv(resFile, sep=';')
```

```python
# Remove neutral group (they did not do the mirror task)
df = df.loc[df['type_modif'] != 0]
```

```python
df.head()
```

```python
len_bb = df.groupby(["bb_nb",'phase'])["key_name"].count().reset_index()
len_bb = len_bb.rename(columns={"key_name": "tot_key"})
len_bb.head()
```

```python
count_keys = df[df["key_name"].isin(["Mi"])]
count_keys = count_keys.groupby(["bb_nb",'phase','key_name'])["Unnamed: 0"].count().reset_index()
count_keys = count_keys.rename(columns={"Unnamed: 0": "Mi_nb"})
count_keys.head()
```

```python
## Get the % of time spent looking at the mirror = Compute the % of "Mi" in key_name column
## number of "Mi" / len["key_name"] x100
mirror = pandas.merge(count_keys, len_bb, on = ["bb_nb", "phase"])
#mirror.head()
mirror["mi_percentage"] = (mirror["Mi_nb"] / mirror["tot_key"])*100
mirror.head(10)
```

```python
plt.figure(figsize=[15,3])
sns.barplot(x = "bb_nb", hue="phase", y="mi_percentage", data= mirror, palette = "BuGn")
sns.despine()
plt.axhline(y=0, linewidth=0.5, color = 'k', linestyle = "--")
plt.xticks(fontsize=8, rotation=45)
plt.legend(frameon=False)
```

```python
# Compute Mi average per phase (before vs after mark)
mirror1 = mirror.loc[mirror["phase"] == "mirror1"]
mean_mirror1 = mirror1["mi_percentage"].mean()
std_mirror1 = mirror1["mi_percentage"].std()
print(f"moyenne du temps de regard vers le miroir en phase 1 : {mean_mirror1} avec une std de {std_mirror1}")

# Select phase 2 + exclude bb101 & 103 , who are uncodable for almost all their phase 1: we can't make the comparisn for them
mirror2 = mirror.loc[
    (mirror["phase"] == "mirror2") & (~mirror["bb_nb"].isin([101, 103]))
]
mean_mirror2 = mirror2["mi_percentage"].mean()
std_mirror2 = mirror2["mi_percentage"].std()
print(f"moyenne du temps de regard vers le miroir en phase 2 : {mean_mirror2} avec une std de {std_mirror2}")

print(len(mirror1)) 
print(len(mirror2))
```

```python
print("test the difference", stats.ttest_ind(mirror1["mi_percentage"], mirror2["mi_percentage"]))
```

```python
## % of time in Tm+Ta = Calculer les pourcentage de "Tm" & "Ta" dans la colonne key_name
## Nombre de "Tm" & "Ta" / len["key_name"] x100
touch_mi = df[df["key_name"].isin(["Tm", "Ta"])]
touch_mi = touch_mi.groupby(["bb_nb",'phase'])["Unnamed: 0"].count().reset_index()
touch_mi = touch_mi.rename(columns={"Unnamed: 0": "Touch/reach_nb"})


## Nombre de "Tm" & "Ta" / len["key_name"] x100
touch_mi = pandas.merge(touch_mi, len_bb, on = ["bb_nb", "phase"])
touch_mi["Tm_percentage"] = (touch_mi["Touch/reach_nb"] / touch_mi["tot_key"])*100
touch_mi.head(10)
```

```python
plt.figure(figsize=[10,3])
sns.barplot(x = "bb_nb", hue="phase", y="Tm_percentage", data= touch_mi, palette = "BuGn")
sns.despine()
plt.axhline(y=0, linewidth=0.5, color = 'k', linestyle = "--")
plt.xticks(fontsize=8, rotation=45)
plt.legend(frameon=False)
```
