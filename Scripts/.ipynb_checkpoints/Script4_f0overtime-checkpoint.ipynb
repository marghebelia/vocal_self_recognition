## Variance of f0 (Supplementary analysis 4)

```python
# pitch_variation = standard deviation of pitch / mean pitch * 100
df_m = df_bb.groupby(["BB", "cond", "phase"])["f0_BBvoc_NM"].mean().reset_index(name = 'mean_pitch')
df_std = df_bb.groupby(["BB", "cond", "phase"])["f0_BBvoc_NM"].std().reset_index(name = "std_pitch")
df_m = pd.merge(df_m, df_std, on = ["BB", "cond", "phase"])
df_m["pitch_variation"] = df_m["std_pitch"] / df_m["mean_pitch"] * 100

## TO EXCLUDE NEUTRAL CONDITION
#df_m = df_m.loc[df_m["cond"] != "neutre"]
```

```python
## PLOTS
plt.figure(figsize=[5,5])
ax = sns.pointplot(x = "phase", hue="cond", y="std_pitch", data= df_m, palette = condColor, units = "BB")
for line in ax.lines:
    line.set_alpha(0.7)
plt.ylabel("pitch variance (SD)")
handles, labels = plt.gca().get_legend_handles_labels()
plt.legend(handles, ["upward", "downward", "neutral"], frameon=False, loc="lower left", fontsize=10)
sns.despine()

# ANOVA INTERACTION CONDITION/phase
model = ols('std_pitch ~ phase + cond + phase:cond', data=df_m).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table, '\n')

p100 = df_m.loc[df_m["cond"] == "p100"]
m100 = df_m.loc[df_m["cond"] == "m100"]
neut = df_m.loc[df_m["cond"] == "neutre"]
print("avg difference in p100 (upward cond):", np.mean(p100["std_pitch"]), np.std(p100["std_pitch"]))
print("avg difference in m100 (downward cond):", np.mean(m100["std_pitch"]), np.std(m100["std_pitch"]))
print("avg difference in neutral condition:", np.mean(neut["std_pitch"]), np.std(neut["std_pitch"]))
print("test the p100 vs m100", stats.ttest_ind(p100["std_pitch"], m100["std_pitch"]))
print("test the neut vs p100", stats.ttest_ind(p100["std_pitch"], neut["std_pitch"]))
print("test the neut vs m100", stats.ttest_ind(neut["std_pitch"], m100["std_pitch"]), "\n")

# Enregistrer le graphique dans un fichier
filename = f"FIGS4_std_pitch.png"
filepath = os.path.join(output_dir, filename)
plt.savefig(filepath, dpi = 300)

filename_pdf = f"FIGURES4_std_pitch.pdf"
filepath_pdf = os.path.join(output_dir, filename_pdf)
plt.tight_layout()
plt.savefig(filepath_pdf)
```

## FO OVER TIME PER CONDITION?

```python
nboot = 100
maxtime = 2
#df_bb["time"] = df_bb["timepoint"]; # use all timestamps
df_bb["time"] = np.round(df_bb["timepoint"]*5, 1)/5; # round up to 20ms
#df_bb["time"] = np.round(df_bb["timepoint"]*2, 1)/2; # round up to 50ms
#df_bb["time"] = np.round(df_bb["timepoint"], 1); # round up to 100ms
#df_bb["time"] = np.round(df_bb["timepoint"]/2, 1)*2; # round up to 200ms
#df_bb["time"] = np.round(df_bb["timepoint"]/5, 1)*5; # round up to 500ms
print(len(np.unique(df_bb["BB"])))

df["M_minus_NM"] = df["f0_BBvoc_M"] - df["f0_BBvoc_NM"] # 4 BB
df_m = df_bb.groupby(["BB", "time", "phase", "cond", "start_times"])[["f0_BBvoc_NM", "f0_BBvoc_M", "M_minus_NM"]].mean().reset_index()

## COMPUTES Z-SCORE / BASELINE for each vocalisation in the NM files for BB vocs
baseline = df_m.loc[df_m["phase"] == "BL"]  # selectionne juste la baseline
baseline_m = baseline.groupby(["BB", "cond"])[["f0_BBvoc_NM", "f0_BBvoc_M"]].mean().reset_index()
baseline_m = baseline_m.rename(columns={"f0_BBvoc_NM": "baseline_mean_f0_NM", "f0_BBvoc_M": "baseline_mean_f0_M"})
df_m = pd.merge(df_m, baseline_m, on = ["BB", "cond"])
baseline_sd = baseline.groupby(["BB", "cond"])[["f0_BBvoc_NM", "f0_BBvoc_M"]].std().reset_index()
baseline_sd = baseline_sd.rename(columns={"f0_BBvoc_NM": "baseline_std_f0_NM", "f0_BBvoc_M": "baseline_std_f0_M"})
df_m = pd.merge(df_m, baseline_sd, on = ["BB", "cond"])
df_m["zf0_Original"] = (df_m["f0_BBvoc_NM"] - df_m["baseline_mean_f0_NM"]) / df_m["baseline_std_f0_NM"]
df_m = df_m.loc[df_m["time"] <= maxtime] ## keeps only timepoints < 3 seconds in the analysisL...
df_m_m = df_m.groupby(["BB", "time", "phase", "cond"])[["f0_BBvoc_NM", "f0_BBvoc_M", "M_minus_NM","zf0_Original"]].mean().reset_index()

if excl_neutral:
    df_m_m = df_m_m.loc[df_m_m["cond"] != "neutral"] # if you want to exclude neutral

print(len(np.unique(df_m_m["BB"])))
np.unique(df_m_m["time"])
```
