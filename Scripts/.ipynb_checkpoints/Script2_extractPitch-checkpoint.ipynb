```python
### IMPORT PACKAGES
import matplotlib.pylab as plt
import numpy as np
import os, glob
import pandas as pd
import seaborn as sns
import re
from scipy import stats, signal
from scipy.io import wavfile
import statsmodels.api as sm
from statsmodels.formula.api import ols
from utilities_SOA import Extract_ts_of_pitch_praat, get_HNR
```

```python
## DATAFRAME CREATION
dft = pd.DataFrame({}) # initialise the f0 * voc dataframe
dfHNR_overtime = pd.DataFrame({}) 
df_all = pd.read_csv(nameresfile, sep = ";")
df_all = df_all.drop(columns=['Unnamed: 0'])
df_all.head()
```

### Loop to extract the f0 and compute our main variables 

```python
### LIST SUBJECT NUMBER
df_all = df_all[~np.isnan(df_all["BB"])]
df_all[["BB"]] = df_all[["BB"]].astype(int)
bb_list = np.unique(df_all["BB"])
print(" N =", len(bb_list), bb_list)
```

```python
### MAIN LOOP FOR EXTRACTING F0 BY BB BY SOUND
for ix, row in df_all.iterrows(): ## for every row
    if row['who'] in ['N', 'CoV', 'UnS']: #skip pitch extraction for N and CoV
        continue

#    if row['BB'] != 1:
#        continue

    audiofile_NM = DIR_SOUNDS + str(row["BBvoc_NM"]) + ".wav"
    audiofile_M = DIR_SOUNDS + str(row["BBvoc_M"]) + ".wav"
    print(audiofile_NM, audiofile_M)

    # f0 extraction from NM and M files
    t, f0_NM = Extract_ts_of_pitch_praat(audiofile_NM)
    t, f0_M = Extract_ts_of_pitch_praat(audiofile_M)
    
    ### CREATES A DATAFRAME WITH ALL DATAPOINTS
    dft_voc = pd.DataFrame({
                        'file': np.tile(row["file"],len(t)),
                        'start_times': np.tile(row["start_times"],len(t)),
                        'stop_times': np.tile(row["stop_times"],len(t)),
                        'who': np.tile(row["who"],len(t)),
                        'cond': np.tile(row["cond"],len(t)),
                        'voc_type': np.tile(row["voc_type"],len(t)),
                        'phase': np.tile(row["phase"],len(t)),
                        'BBvoc_NM':np.tile(row["BBvoc_NM"],len(t)),
                        'BBvoc_M':np.tile(row["BBvoc_M"],len(t)),
                        'BB':np.tile(row["BB"],len(t)),
                        "f0_BBvoc_NM": f0_NM,
                        "f0_BBvoc_M": f0_M,
                        'timepoint': t})
    
    ### HNR & RMS extraction
    signals = {}
    rates = {}

    rate1, signal1 = wavfile.read(audiofile_NM)
    rate2, signal2 = wavfile.read(audiofile_M)

    # HNR 
    HNR_NM = get_HNR(signal1,rate1)
    HNR_M = get_HNR(signal2,rate2)
    
    # amplitude extraction
    RMS_NM = np.mean(np.abs(signal.hilbert(signal1)))
    RMS_M = np.mean(np.abs(signal.hilbert(signal2)))

    df_HNR_RMS_overtime = pd.DataFrame({
                        'file': np.tile(row["file"],1),
                        'start_times': np.tile(row["start_times"],1),
                        'stop_times': np.tile(row["stop_times"],1),
                        'who': np.tile(row["who"],1),
                        'cond': np.tile(row["cond"],1),
                        'voc_type': np.tile(row["voc_type"],1),
                        'phase': np.tile(row["phase"],1),
                        'BBvoc_NM':np.tile(row["BBvoc_NM"],1),
                        'BBvoc_M':np.tile(row["BBvoc_M"],1),
                        'BB':np.tile(row["BB"],1),
                        "HNR_NM": HNR_NM,
                        "HNR_M": HNR_M,
                        "RMS_NM": RMS_NM,
                        "RMS_M": RMS_M})
    
    ## ADD NEW DATA TO THE SEQUENCE following DF f0 and DF hnr_rms
    dft = pd.concat([dft, dft_voc], ignore_index=True)
    dfHNR_overtime = pd.concat([dfHNR_overtime, df_HNR_RMS_overtime], ignore_index=True)

    # clean memory 
    del f0_NM, f0_M, HNR_NM, HNR_M, RMS_NM, RMS_M
```

### Compute f0 difference between the modified and unmodified sounds 

```python
dtype_dict = {6: str}

#Compute f0 difference
dft['difference'] = dft['f0_BBvoc_M'] - dft['f0_BBvoc_NM']

# Put NaN if f0 difference is > 250Hz
dft.loc[(dft['difference'] > 250) | (dft['difference'] < -250), ['difference', 'f0_BBvoc_M', 'f0_BBvoc_NM']] = np.nan

#3) Put NaN if f0_NM & f0_M are the same value
dft.loc[pd.isna(dft['difference']), ['f0_BBvoc_M', 'f0_BBvoc_NM']] = np.nan

## SAVES df to csv
dft.to_csv(DIR_data + "N" + str(len(bb_list)) + "_f0_overtime.csv", sep = ";")
dfHNR_overtime.to_csv(DIR_data + "N" + str(len(bb_list)) + "_HNR_RMS_overtime.csv", sep = ";")
```

```python
## LOAD FILE 

# Compute f0_NM
dft['f0m_NM'] = dft.groupby(['start_times', 'file', 'who', "BB"])['f0_BBvoc_NM'].transform('mean')
dft['f0sd_NM'] = dft.groupby(['start_times', 'file', 'who', "BB"])['f0_BBvoc_NM'].transform('std')
dft['f0pNa_NM'] = dft.groupby(['start_times', 'file', 'who', "BB"])['f0_BBvoc_NM'].transform(lambda x: x.isna().mean())
dft['f0c_NM'] = dft.groupby(['phase', 'file', 'who', "BB"])['f0_BBvoc_NM'].transform('mean')
dft['f0cSD_NM'] = dft.groupby(['phase', 'file', 'who', "BB"])['f0_BBvoc_NM'].transform('std')

# Compute f0_M
dft['f0m_M'] = dft.groupby(['start_times', 'file', 'who', "BB"])['f0_BBvoc_M'].transform('mean')
dft['f0sd_M'] = dft.groupby(['start_times', 'file', 'who', "BB"])['f0_BBvoc_M'].transform('std')
dft['f0pNa_M'] = dft.groupby(['start_times', 'file', 'who', "BB"])['f0_BBvoc_M'].transform(lambda x: x.isna().mean())
dft['f0c_M'] = dft.groupby(['phase', 'file', 'who', "BB"])['f0_BBvoc_M'].transform('mean')
dft['f0cSD_M'] = dft.groupby(['phase', 'file', 'who', "BB"])['f0_BBvoc_M'].transform('std')

# We want that dft has the same number of rows of df_all (1 row per voc).
dft = dft[['file', 'start_times', 'stop_times', "BB", "who", 'f0m_NM', 'f0sd_NM', 'f0pNa_NM', 'f0c_NM',
           'f0m_M', 'f0sd_M', 'f0pNa_M', 'f0c_M', "f0cSD_NM", "f0cSD_M"]]
dft = dft.drop_duplicates()
print(len(dft), "len dft")
print(len(df_all), "len df all b4 merge with dft")
df_all = df_all.merge(dft)
#df_all = df_all.drop(columns = ['index','Unnamed: 34', 'coded', 'audio_order', 'category1_M','category2_M', 'category1_F', 'category2_F', 'agr_cat1', 'agr_cat2'])
df_all = df_all.drop_duplicates()
print(len(df_all), "len of dfall after merging mean pitch")

### SAME FOR RMS
# Compute RMS_NM (mesure of the amplitude signal, linked to the perception of the sound in dB)
dfHNR_overtime['RMSm_NM'] = dfHNR_overtime.groupby(['start_times', 'file', 'who', "BB"])['RMS_NM'].transform('mean')
dfHNR_overtime['RMSsd_NM'] = dfHNR_overtime.groupby(['start_times', 'file', 'who', "BB"])['RMS_NM'].transform('std')
#dfHNR_overtime['RMSpNa_NM'] = dfHNR_overtime.groupby(['start_times', 'file'])['RMS_NM'].transform(lambda x: x.isna().mean())

# Compute RMS_M
dfHNR_overtime['RMSm_M'] = dfHNR_overtime.groupby(['start_times', 'file', 'who', "BB"])['RMS_M'].transform('mean')
dfHNR_overtime['RMSsd_M'] = dfHNR_overtime.groupby(['start_times', 'file', 'who', "BB"])['RMS_M'].transform('std')
#dfHNR_overtime['RMSpNa_M'] = dfHNR_overtime.groupby(['start_times', 'file'])['RMS_M'].transform(lambda x: x.isna().mean())

## MERGE
dfHNR_overtime = dfHNR_overtime[['file', 'start_times', 'stop_times', 'who', "BB", 'HNR_M', 'HNR_NM', 
                                 'RMSm_NM', 'RMSsd_NM', 'RMSm_M', 'RMSsd_M']] #, 'RMSpNa_NM', 'RMSpNa_M'
dfHNR_overtime = dfHNR_overtime.drop_duplicates()
df_all = df_all.merge(dfHNR_overtime)
df_all = df_all.drop_duplicates()
print(len(df_all), "len of dfall after merging mean HNR AND RMS")

## Save df
df_all.to_csv(DIR_data_git + "N" + str(len(bb_list)) + "_f0_HNR_RMS.csv", sep = ";", index=False)
```
